```python
import matplotlib.pyplot as plt
import random
import math
import numpy as np
```

```python
def plot_ifs_fractal(coefficients, probabilities, n_points=100000, color='black', title='IFS Fractal'):
    x, y = 0, 0
    x_vals = []
    y_vals = []

    for _ in range(n_points):
        a, b, c, d, e, f = random.choices(coefficients, weights=probabilities)[0]
        x_new = a * x + b * y + e
        y_new = c * x + d * y + f
        x, y = x_new, y_new
        x_vals.append(x)
        y_vals.append(y)

    plt.figure(figsize=(6, 6))
    plt.scatter(x_vals, y_vals, s=0.1, color=color)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

5. Модифікувати програмне забезпечення з 4-го пункту завдань до даної
лабораторної роботи динамічною (по ітераціях) візуалізацією кожного
фракталу. Для деяких фракталів потрібно буде виводити не кожну точку
ітерації, а кілька точок згенерованих протягом M (задає розробник
програми) ітерацій, оскільки для досить великої кількості точок виведення
по одній точці буде тривалим у часі.

```python
def plot_ifs_dynamic(coefficients, probabilities, n_points=5000, batch_size=10000, color='black', title='IFS Fractal'):
    x, y = 0, 0
    xs, ys = [], []

    plt.ion()
    fig, ax = plt.subplots()
    ax.set_title(title)
    ax.axis('off')

    sc = ax.scatter([], [], s=0.1, color=color)

    for i in range(n_points):
        a, b, c, d, e, f = random.choices(coefficients, weights=probabilities)[0]
        x_new = a * x + b * y + e
        y_new = c * x + d * y + f
        x, y = x_new, y_new
        xs.append(x)
        ys.append(y)

        if (i + 1) % batch_size == 0:
            sc.set_offsets(np.column_stack((xs, ys)))
            # Автоматично підлаштовуємо межі
            ax.set_xlim(min(xs), max(xs))
            ax.set_ylim(min(ys), max(ys))
            plt.pause(0.001)

    sc.set_offsets(np.column_stack((xs, ys)))
    ax.set_xlim(min(xs), max(xs))
    ax.set_ylim(min(ys), max(ys))
    plt.ioff()
    plt.show()
```

```python
plot_ifs_dynamic(maple_leaf_coeffs, maple_leaf_probs, n_points=50000, batch_size=10000, color='green', title='Кленовий лист')
```
